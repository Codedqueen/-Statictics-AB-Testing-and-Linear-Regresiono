#### Confidence Interval (CI)

This indicates how much uncertainty there is in a Sample Statistic(such as mean) of a population pararmeter.

They are computed to quantify how confident we can be in our results from our survey/sample observations/sample experiment that these reflects,up to what extent this reflects the entire population.

95% CI has been used very widely accross different fields to asses how much viariability will appear in repeated measures.

95% CI means if we repeat an experiment again/take multiple samples, observations or studies and compute the confidence interval for each sample,95% of the intervals would contain the population mean

CI computation involves computing the sample statistics examle, sample meanm sample proportionand margin of error(MOE) which  expresses the amount of random sampling error in the sample results.

Especially when we are working with different samples and we dont have census data, there is going to be random sampling error which will be accounted for.

#### Calculate the confidence interval in Python

Point estimate are estimates of population parameters based on sample data.

If we want to know the average weight of students in a school, we will collect a sample to estimate the average weight of all the students in the school.

The sample mean is usually not exactly thesame as the population mean.

A cofidance interval is a range of values above and below a point estimate that captures the true population parameter at some predetermined confidence level.

If you want to have a 95% chance of capturing the true population parameter with a point estimate and a corresponding confidence interval, you will set your confidence level at 95%

CI = mean +- margin of error 

where MOE ie Z*sd /sqrt(n)

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
import math
```

```python
np.random.seed(10)

population_wt = stats.poisson.rvs(loc=18, mu=35, size=150000)
#randomly create population weight data for 150k people

sample_size = 1000  # sample of 1000 people
sample = np.random.choice(a = population_wt, size =sample_size) # Create a sample weight
sample_mean = sample.mean() # create a sample mean

# There is no guarantee that sample mean will be the same as population mean

z_critical = stats.norm.ppf(q = 0.975)
# get the z critical value of 95% level
#z value at 95% CI = 1.96

print("z-critical value:")
print(z_critical)  # check the z-critical value

pop_stdev = population_wt.std()
#get the population standard deviation)

margin_of_error = z_critical * (pop_stdev / math.sqrt(sample_size)) # To calculate margin od error

confidence_interval = (sample_mean - margin_of_error, 
                       sample_mean + margin_of_error) # To find confidence interval

print("confidence interval:")
print(confidence_interval)
```
